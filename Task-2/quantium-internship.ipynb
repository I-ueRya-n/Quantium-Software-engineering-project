```python
# Import necessary libraries
import pandas as pd

# Load the CSV files
# Load the CSV files
df_2018 = pd.read_csv('/kaggle/input/daily-sales-data/daily_sales_data_0.csv')
df_2019 = pd.read_csv('/kaggle/input/daily-sales-data/daily_sales_data_1.csv')
df_2020 = pd.read_csv('/kaggle/input/daily-sales-data/daily_sales_data_2.csv')

# Print the first few rows of each DataFrame to inspect
print("Data from 2018 file:")
print(df_2018.head())

print("Data from 2019 file:")
print(df_2019.head())

print("Data from 2020 file:")
print(df_2020.head())

# Inspect unique product names in each DataFrame
print("Unique products in 2018 file:")
print(df_2018['product'].unique())

print("Unique products in 2019 file:")
print(df_2019['product'].unique())

print("Unique products in 2020 file:")
print(df_2020['product'].unique())

# Standardize product names by stripping spaces and converting to lower case
df_2018['product'] = df_2018['product'].str.strip().str.lower()
df_2019['product'] = df_2019['product'].str.strip().str.lower()
df_2020['product'] = df_2020['product'].str.strip().str.lower()

# Filter rows where the product is 'pink morsel'
df_2018_filtered = df_2018[df_2018['product'] == 'pink morsel']
df_2019_filtered = df_2019[df_2019['product'] == 'pink morsel']
df_2020_filtered = df_2020[df_2020['product'] == 'pink morsel']

# Print the number of rows after filtering
print("Rows with 'pink morsel' in 2018 file:", len(df_2018_filtered))
print("Rows with 'pink morsel' in 2019 file:", len(df_2019_filtered))
print("Rows with 'pink morsel' in 2020 file:", len(df_2020_filtered))

# Convert price to float after stripping the dollar sign using .loc
df_2018_filtered.loc[:, 'price'] = df_2018_filtered['price'].replace('[\$,]', '', regex=True).astype(float)
df_2019_filtered.loc[:, 'price'] = df_2019_filtered['price'].replace('[\$,]', '', regex=True).astype(float)
df_2020_filtered.loc[:, 'price'] = df_2020_filtered['price'].replace('[\$,]', '', regex=True).astype(float)

# Calculate the sales (quantity * price)
df_2018_filtered['sales'] = df_2018_filtered['quantity'] * df_2018_filtered['price']
df_2019_filtered['sales'] = df_2019_filtered['quantity'] * df_2019_filtered['price']
df_2020_filtered['sales'] = df_2020_filtered['quantity'] * df_2020_filtered['price']

# Select only the necessary columns (sales, date, region)
df_2018_filtered = df_2018_filtered[['sales', 'date', 'region']]
df_2019_filtered = df_2019_filtered[['sales', 'date', 'region']]
df_2020_filtered = df_2020_filtered[['sales', 'date', 'region']]

# Combine the dataframes into one
combined_df = pd.concat([df_2018_filtered, df_2019_filtered, df_2020_filtered])

# Print the combined dataframe to inspect
print("Combined DataFrame:")
print(combined_df.head(10))

# Save the combined dataframe to a new CSV file
combined_df.to_csv('combined_sales_data.csv', index=False)
```
